```python
import re
from docx import Document

def get_text_from_docx(filename):
    """Extrahiert den gesamten Text aus einem .docx-Dokument."""
    doc = Document(filename)
    return '\n'.join([para.text for para in doc.paragraphs])

def clean_and_format_text(text_to_clean):
    """Führt alle Textbereinigungen und Ersetzungen durch."""
    # Ersetzung von Wörtern und Begriffen
    replacements = {
        "make space": "Makerspace",
        "Journis": "Journals",
        "uni": "Universität",
    }
    for old, new in replacements.items():
        text_to_clean = text_to_clean.replace(old, new)

    # Unerwünschte Ausdrücke entfernen, hier mit RegEx für 'ähm' und 'Ähm'
    text_to_clean = re.sub(r'\s*(ähm|Ähm)\s*', ' ', text_to_clean)

    # Entfernen von Zeitstempeln wie #00:00:09-1#
    text_to_clean = re.sub(r'#\d{2}:\d{2}:\d{2}-\d#', '', text_to_clean)

    # Reduzieren von mehrfachen Leerzeichen zu einem einzigen Leerzeichen und entfernen von Leerzeichen am Anfang und Ende
    cleaned_text = re.sub(r'\s+', ' ', text_to_clean).strip()
    
    return cleaned_text

# --- Hauptprogramm ---
# 1. Text aus der Word-Datei extrahieren
file_path = '../projekt/data/HyLeC.docx'
raw_document_text = get_text_from_docx(file_path)

# 2. Den extrahierten Text bereinigen
final_cleaned_text = clean_and_format_text(raw_document_text)

# 3. Das Ergebnis ausgeben
print(final_cleaned_text)
```
