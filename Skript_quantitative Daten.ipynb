## Fehlende Werte identifizieren und behandeln

```python
# Möchte man genau wissen, wie viele Werte fehlen, kann man folgenden Code nutzen.

fehlende_Werte = df.isna().sum()
print(fehlende_Werte) 

# Achtung: mit diesem Code wird der DataFrame zu einem Series. Daher wird eine neue Variable eingeführt, um beide Version zu zeigen.
```

```python
# Zur Bereinigung wird folgender Code verwendet: 1. Variante

bereinigte_Werte = df.dropna()
bereinigte_Werte.isna().sum()
```

```python
# 2. Variante: Alle Spalten entfernen, die keine Werte enthalten. Danach alle Zeilen entfernen, die keine Werte enthalten.
# "all" betrifft alle SPalten bzw. Zeilen, die keine Werte entghalten. 0 betrifft die Zeilen, 1 die Spaltem.

#df_bereinigt = df.dropna(how="all", axis =1)
#df_bereinigt = df.dropna(how="all", axis =0)

#Zusammengefügter Code:
df_bereinigt = df.dropna(how="all", axis=1).dropna(how="all", axis=0)
```

```python
df_bereinigt
```

In diesem Beispiel sind keine vollständig leere Spalten oder Zeilen vorhanden. Daher wurde hier nichts gelöscht.
Für die weitere Betrachtung wird mit bereinigte_Werte gearbeitet.

# Spalten löschen, die nicht benötigt werden.
In diesem Beispiel enthalten die Spalten _fact_key_ und institution dieselben Werte.

```python
# Entfernen einer Spalte. Hier:_fact_key-Spalte wird entfernt.

df_clear = bereinigte_Werte.drop(columns=['__fact_key__'])
```

```python
# Um zu prüfen, ob eine Spalte dieselben Werte ausgibt, wird eine if-Bedingung genutzt.
# Nunique gibt die Anzahl der eindeutigen Werte zurück, die in der Spalte "Country" vorhanden sind.
# Sind ungleiche Werte vorhanden, wird "unterschiedliche Werte" als Ergebnis ausgegeben.

if df_clear['country'].nunique() == 1:
    print("Gleiche Werte")
else:
    print("Unterschiedliche Werte")
```

```python
# Da die Spalte nur "DEU"-Werte hat, wird sie ebenso gelöscht. Das alte DataFrame "df_clear" wird überschrieben.

df = df_clear.drop(columns=['country'])
```

```python
# Überprüfung
df
```

# Werte bereinigen
1. Umlaute
2. Sonderzeichen
3. Datumformat
4. Umbennung

```python
# 1. Umlaute anpassen und vereinheitlichen. Hier z.B. Fachhochschule SÃ¼dwestfalen
# mit der replace-Funktion die werden fehlerhaften Buchstaben ersetzt.

df['institution'] = df['institution'].str.replace('Ã¼', 'ue')
df['institution'] = df['institution'].str.replace('A¤', 'ae')
df['institution'] = df['institution'].str.replace('A¶', 'oe')

df

# Achtung: Diese Variante umfasst nur die Behebung in einer Spalte.
```

```python
# Sonderzeichen entfernen mit replace-Funktion. Hier: Bei Spalte "Price_eur" ";" durch nichts.

df['price'] = df['price'].str.replace(';', '')
df
```

```python
# Sonderzeichen hinzufügen. 
# Dafür muss die Spalte als String umgewandelt werden mit .astype und das Sonderzeichen hinzugefügt werden (mit Leerzeichen)

df['price'] = df['price'].astype(str) + ' €'
df
```

```python
# Datumformat anpassen. Hier in Spalte "year". Dafür muss der Datentyp zu Integer geändert werden.

df['year'] = df['year'].astype(int)
df
```

```python
# 4. Umbenennung von Werten. Hier: true zu hybrid und false zu not-hybrid.

df['hybrid'] = np.where(df['hybrid'] == True, 'hybrid', 'not-hybrid')
df
```

# Metadaten anzeigen und anpassen

```python
# Die Bibliothek os bietet Zugriff auf Betriebssystem-Informationen.Somit lassem sich Datei- und Pfadinformationen abrufen.
# Es muss zunächst der Dateipfad angegeben werden. Danach werden detaillierte Metadaten der Datei abgerufen mit Funktion os.stat

import os

dateipfad = "../projekt/data/20240825_OpenAPC_raw.csv"
dateistatus = os.stat(dateipfad)

# Anzeige Metadaten
print(f"Dateigröße: {dateistatus.st_size} bytes")
print(f"Datum des letzten Zugriffs: {dateistatus.st_atime}")
print(f"Datum der letzten Änderung: {dateistatus.st_mtime}")
```

```python
# Datumangabe anpassen und neue Ausgabe

import time
import datetime

zugriffszeit = datetime.datetime.fromtimestamp(dateistatus.st_atime)
aenderungszeit = datetime.datetime.fromtimestamp(dateistatus.st_mtime)

print(f"Dateigröße: {dateistatus.st_size} bytes")
print(f"Datum des letzten Zugriffs: {zugriffszeit.strftime('%Y-%m-%d')}")
print(f"Datum der letzten Änderung: {aenderungszeit.strftime('%Y-%m-%d')}")
```

```python
# Dictionary für Metadaten erstellen und ausgeben lassen
metadaten_dict = {"Metadaten": ["Dateigröße (Bytes)", "Letzter Zugriff", "Letzte Änderung", "Autorin", "Lizenz", "Publikation"],
"Ausgabe": [dateistatus.st_size, zugriffszeit.strftime('%Y-%m-%d'), aenderungszeit.strftime('%Y-%m-%d'), autorin, lizenz_info, publikation]}

metadaten = metadaten_dict.items()
print(metadaten)
```

```python
# Exkurs: Inhaltsangabe der Tabelle für Überblick der Datenstruktur und Qualitätssicherung (fehlende Werte)

def ueberblickswerte (df):
    
    daten = pd.DataFrame({
        'Variable': df.columns,
        'Datentyp': df.dtypes.astype(str),
        'Anzahl fehlender Werte': df.isnull().sum(),
    })
    
    daten['Beispielwerte'] = [
        f"Min: {df[col].min()}, Max: {df[col].max()}" if pd.api.types.is_numeric_dtype(df[col]) 
        else f"Beispiel: {df[col].iloc[0]}"
        for col in df.columns
    ]
    
    return daten.set_index('Variable')

datenstruktur = ueberblickswerte(df)

print(datenstruktur)
```
