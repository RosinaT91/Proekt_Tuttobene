```python
# Dokument einlesen und kompletten Text abrufen mithilfe einer Funktion, die den Text in der Datei als einzelnen Stringwert zurückgibt.
# Achtung: python-docx muss im Terminal installiert sein (s. README-Datei)

import docx

def getText():
    doc = docx.Document('../projekt/data/HyLeC.docx')
    fulltext = []
    for para in doc.paragraphs:
        fulltext.append(para.text)
    return '\n'.join(fulltext)
```

```python
# Funktion aufrufen, auf Text verwenden und neue Variable für String erstellt.

text = getText()

#Text wird ausgegeben.
print(text)
```

# Textbereinigung

```python
# wichtige Bibliotheken importieren und Text tokenisieren (bessere Lesbarkeit durch Zerlegung des Textes in kleinere Teile)

import nltk
import re
from nltk.tokenize import word_tokenize
```

```python
# Begriffe ersetzen: Liste anlegen (Tupel) mit falschen und richtigen Begriffen und mithilfe einer for-Schleife auf den Text anwenden.
# ähm und Ähm sollen durch nichts ersetzt werden --> Löschung

replace = [
    ("make space", "Makerspace"),
    ("Journis", "Journals"),
    ("uni", "Universität"),
    ("Lerning Center", "Learning Center"),
     ("Lernen ort", "Lernort"),
    ("Futur sci", "Future Skills"),
    ("ähm", ""), 
    ("Ähm", "")
]

for false_term, right_term in replace:
    text = text.replace(false_term, right_term)
```

```python
# Überprüfung des Ergebnisses

print(text)
```

```python
# Ersetzen von Sprechernamen. Hier Speaker 2 zu "Interviewer" und Speaker 1 zu "Befragte Person"

text = re.sub(r'Speaker 2', 'Interviewer', text)
text = re.sub(r'Speaker 1', 'Befragte(r)', text)
```

```python
# Überprüfung des Ergebnisses

print(text)
```

```python
# Doppelte Leerzeichen entfernen

cleaned_text = re.sub(r'\s+', ' ', text).strip()
```

```python
# Überprüfung des Ergebnisses

print(cleaned_text)
```

# Anonymisierung

```python
#Variante 1: manuelle Definition der Begriffe, die anonymisiert werden sollen


replace = [
    ("Hybrid Learning center", "<INSTITUTION>"),
    ("Learning Center", "<INSTITUTION>"),
    ("HyLeC", "<INSTITUTION>"),
    ("hylec", "<INSTITUTION>"),
     # ("Frau xy", "<PERSON")
]

for name, anonym in replace:
    anonymized_text = cleaned_text.replace(name, anonym)

print(anonymized_text)
```

```python
# Variante 2: automatisierte Anonymisierung mit spacy
# Achtung: spacy muss im Terminal installiert sein (s. README-Datei) --> s. Probleme mit numpy-BIbliothek

# Import der Bibliothek sowie des deutschen vortrainierten Sprachmodells. 
import spacy
nlp = spacy.load("de_core_news_sm")

# Doc-Objekt wird erstellt für Text
doc = nlp(cleaned_text)

#For-Schleife
#"PER" = Person
#"LOC" = Location
# Namen und Orte sollen durch PLatzhalter (PER oder LOC) ersetzt werden.

for ent in reversed(doc.ents):
    if ent.label_ in ["PER", "LOC"]:
        cleaned_text = cleaned_text[:ent.start_char] + f"[{ent.label_}]" + cleaned_text[ent.end_char:]

print(cleaned_text)
```

```python
# Problem: "Interviewer" und "Befragter(r) werden u.U. auch ersetzt. Daher. not in"

import spacy
nlp = spacy.load("de_core_news_sm")
doc = nlp(cleaned_text)

for ent in reversed(doc.ents):
    if ent.label_ in ["PER", "LOC"] and ent.text not in ["Interviewer", "Befragte(r)"]:
        cleaned_text = cleaned_text[:ent.start_char] + f"[{ent.label_}]" + cleaned_text[ent.end_char:]

print(cleaned_text)
```
