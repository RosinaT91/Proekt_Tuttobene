# Skript zur automatisierten Aufbereitung qualitativer Daten

```python
# Dokument einlesen und kompletten Text abrufen mithilfe einer Funktion, die den Text in der Datei als einzelnen Stringwert zurückgibt.
# Achtung: python-docx muss im Terminal installiert sein (s. README-Datei)

import docx

def getText():
    doc = docx.Document('../projekt/data/HyLeC.docx')
    fulltext = []
    for para in doc.paragraphs:
        fulltext.append(para.text)
    return '\n'.join(fulltext)
```

```python
# Funktion aufrufen, auf Text verwenden und neue Variable für String erstellt.

text = getText()

#Text wird ausgegeben.
print(text)
```

# Textbereinigung

```python
# wichtige Bibliotheken importieren und Text tokenisieren (bessere Lesbarkeit durch Zerlegung des Textes in kleinere Teile)

import nltk
import re
from nltk.tokenize import word_tokenize
```

## Begriffe ersetzen

```python
# Begriffe ersetzen: Dictionary mit falschen und richtigen Begriffen und mithilfe einer for-Schleife auf den Text anwenden.
# ähm und Ähm sollen durch nichts ersetzt werden --> Löschung
# für bessere Lesbarkeit wurden die einzelen Paare in Zeilen dargestellt.

replace = {
    "make space": "Makerspace",
    "Journis": "Journals",
    "uni": "Universität",
    "Lerning Center": "Learning Center",
    "Lernen ort": "Lernort",
    "Futur sci": "Future Skills",
    "ähm": "",
    "Ähm": ""}

for false_term, right_term in replace.items():
    text = text.replace(false_term, right_term)
```

```python
# Überprüfung des Ergebnisses

print(text)
```

```python
# Ersetzen von Sprechernamen. Hier Speaker 2 zu "Interviewer" und Speaker 1 zu "Befragte Person"

text = re.sub(r'Speaker 2', 'Interviewer', text)
text = re.sub(r'Speaker 1', 'Befragte(r)', text)
```

```python
# Überprüfung des Ergebnisses

print(text)
```

## Sonderzeichen entfernen

```python
# Doppelte Leerzeichen entfernen

cleaned_text = re.sub(r'\s+', ' ', text).strip()
```

```python
# Überprüfung des Ergebnisses

print(cleaned_text)
```

# Anonymisierung

```python
#Variante 1: manuelle Definition der Begriffe, die anonymisiert werden sollen

replace = {
    "Hybrid Learning center": "INSTITUTION",
    "Learning Center": "INSTITUTION",
    "HyLeC": "INSTITUTION",
    "hylec": "INSTITUTION"
    #"Frau xy": "PERSON"
}


for name, anonym in replace.items():
    anonymized_text = anonymized_text.replace(name, anonym)


print(anonymized_text)
```

```python
# Variante 2: automatisierte Anonymisierung mit spacy
# Achtung: spacy muss im Terminal installiert sein (s. README-Datei) --> s. Probleme mit numpy-BIbliothek (virtuelle Umgebung notwendig)

# Import der Bibliothek sowie des deutschen vortrainierten Sprachmodells. 
import spacy
nlp = spacy.load("de_core_news_sm")

# Doc-Objekt wird erstellt für Text
doc = nlp(cleaned_text)

#For-Schleife
#"PER" = Person
#"ORG" = Organisation
# durch Platzhalter (PER oder ORG) ersetzen

for ent in (doc.ents):
    if ent.label_ in ["PER", "ORG"]:
        cleaned_text = cleaned_text[:ent.start_char] + f"[{ent.label_}]" + cleaned_text[ent.end_char:]

print(cleaned_text)
```

# Metadaten anzeigen und erstellen

```python
# Die Bibliothek os bietet Zugriff auf Betriebssystem-Informationen.Somit lassem sich Datei- und Pfadinformationen abrufen.
# Es muss zunächst der Dateipfad angegeben werden. Danach werden detaillierte Metadaten der Datei abgerufen mit Funktion os.stat

import os

file_path = "../projekt/data/HyLeC.docx"
file_status = os.stat(file_path)

# Anzeige Metadaten
print(f"Dateigröße: {file_status.st_size} bytes")
print(f"Datum des letzten Zugriffs: {file_status.st_atime}")
print(f"Datum der letzten Änderung: {file_status.st_mtime}")
```

```python
# Datumangabe anpassen und neue Ausgabe

import time
import datetime

access_time = datetime.datetime.fromtimestamp(file_status.st_atime)
change_time = datetime.datetime.fromtimestamp(file_status.st_mtime)

print(f"Dateigröße: {file_status.st_size} bytes")
print(f"Datum des letzten Zugriffs: {access_time.strftime('%Y-%m-%d')}")
print(f"Datum der letzten Änderung: {change_time.strftime('%Y-%m-%d')}")
```

# Exkurs: Textanalyse

- Wortwolke erstellen für Überblick
- Hierfür müssen zunächst die Stoppwörter entfernt werden

```python
# Download der Pakete zur Entfernung von Stoppwörtern

from nltk.corpus import stopwords
nltk.download ('stopwords')
nltk.download('punkt_tab')
```

## Stoppwörter

```python
# Anwendung des deutschen Sets von Stoppwörtern
# Definition der Variable tokens (Zerlegung der Wörter)
# Anwendung der Funktion auf bereinigten Text und Erstellung einer neuen Variable

stop_words = set(stopwords.words('german'))
tokens = word_tokenize(cleaned_text)
text_filtered = [token for token in tokens if token not in stop_words]

print(text_filtered)
```

# Stemmatisierung (Reduktion der Wörter auf Stammform)

```python
# Import der notwendigen Werkzeuge und Zuordnung von Variablen

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
# Durchführung
# Liste von Wörtern (Tokens) durchlaufen

stemmed_tokens = [stemmer.stem(token) for token in tokens]
```

```python
# Anwendung auf Text

text_filtered = [stemmer.stem(token) for token in tokens if token not in stop_words]
```

## Wortwolke erstellen

```python
# Import der Bibliothek und auch von matplotlib zur Darstellung

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Alle Wörter aus text_filtered werden zu einem einzigen String zusammengefügt und durch Leerzeichen getrennt.
text_wordcloud = " ".join(text_filtered)

# Anwendung der Methode

wordcloud = WordCloud().generate(text_wordcloud)

# Darstellung der Wortwolke.Parameter interpolation="bilinear": Weichzeichnung der Kanten des Bildes
plt.imshow(wordcloud, interpolation="bilinear")
plt.show()
```

```python
# Da immer noch viele Füllwörter vorhanden sind, werden diese definiert und mit -split gelöscht

no_usage_terms = "und da ja quasi wir gut bisschen eben immer die der das natürlich mal gerade also schon ähm mehr einfach"
split_terms = set(no_usage_terms.split())

# Anwendung wie oben

text_wordcloud = " ".join(text_filtered)
wordcloud = WordCloud(stopwords=split_terms).generate(text_wordcloud)

plt.imshow(wordcloud, interpolation="bilinear")
plt.show()
```

# Sentimentanalyse

```python
# Vorherige Installation der Bibliothek notwendig

from textblob import TextBlob
```

```python
# Analyse
# Polarity= positiver oder negativer Text (Werte: -1 bis 1)
# Subjectivity = subjektiver Inhalt (0 bis 1)
blob = TextBlob(cleaned_text)
sentiment_polarity = blob.sentiment.polarity
sentiment_subjectivity = blob.sentiment.subjectivity
print(f"Sentiment Polarity: {sentiment_polarity}")
print(f"Sentiment Subjectivity: {sentiment_subjectivity}")
```

### Ergebnis: 
Text ist sehr positiv, da bei 0,07 und eher subjektiv.
